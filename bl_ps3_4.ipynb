```python
#!/usr/bin/env python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
file = input('Enter csv file name: ') 
try:
    sample_data = pd.read_csv(file,usecols=[0,1]) #only get columns of interest
    sample_data.columns=sample_data.columns.str.strip().str.replace(' ', '_') #rename \s with underscores
    x = sample_data.Date_Time
    y = sample_data.Water_Level
    plt.plot(x,y)
    plt.title("Water Levels Over Time")
    plt.xlabel("Time")
    plt.ylabel("Water Levels")
    plt.legend(["water levels"])
    plt.show()
except:
    print('File cannot be opened, maybe you didn\'t add \'.csv\'?:', file)
    exit()
```
